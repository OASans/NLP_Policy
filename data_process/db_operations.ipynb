### 政策文件表policy

```python
# 政策uid、原本的政策编号origin_id、政策源文件名policy_name、政策标注状况annotate_status、政策重复状况duplicated_status、标注者名字tagger_name
# annotate_status: 1=annotated;0=unannotated;-1=no_need_to_be_annotated;-2=duplicated
cursor.execute("""create table policy (
uid varchar(20) primary key,
origin_id varchar(20), 
policy_name varchar(20),
annotate_status int,
tagger_name varchar(20),
annotate_time varchar(20)
)""")
```

```python
# 遍历硬盘中的政策文件，每个政策文件都在表中插入一条数据
import re
import glob
folder_list = glob.glob('/Volumes/TOURO Mobil/policy标注/政策文件rawdata/*')
for folder in folder_list:
    uid = folder.split('/')[-1]
    origin_file_list = glob.glob('/Volumes/TOURO Mobil/policy标注/政策文件rawdata/{}/*'.format(uid))
    origin_file_list = [file for file in origin_file_list if (file.split('.')[-1] != 'txt' and file.split('.')[-1] != 'xlsx')]
    if len(origin_file_list) != 1:
        print(uid)
    else:
        file_name = origin_file_list[0].split('/')[-1]
        origin_id = re.split('[._]', file_name)[-2]
        cursor.execute("INSERT INTO policy (uid,origin_id,policy_name,annotate_status,tagger_name, annotate_time) \
                      VALUES ('{}', '{}', '{}', 0, '', '')".format(uid, origin_id, file_name))
conn.commit()
```

### 标注句子表annotated_sentence

```python
# 政策uid、句子sid、句子内容sentence、句子类型sentence_type
cursor.execute("""create table annotated_sentence (
uid varchar(20),
sid varchar(20) primary key,

sentence text,
sentence_type varchar(20)
)""")
```

### 实体信息表entity

```python
# 句子sid、句子中的实体内容entity、实体对应的类别entity_type
cursor.execute("""create table entity (
sid varchar(20),
entity text,
entity_type varchar(20)
)""")
```

### 准入条件表entry

```python
# 句子sid、句子中的准入条件编号eid、准入条件类别、变量、关系、域、变量原文、域原文
cursor.execute("""create table entry (
sid varchar(20),
eid varchar(20),
entry_type varchar(20),
var text,
relation varchar(20),
field text,
var_context text,
field_context text,
primary key (sid,eid)
)""")
```

### 准入条件逻辑表entry_logic

```python
# 政策uid、准入条件逻辑关系logic、准入条件认定组别logic_name
cursor.execute("""create table entry_logic (
uid varchar(20),
logic varchar(255),
logic_name varchar(255)
)""")
```

## 2 写记录api

```python
# 更新policy表状态
# info = (annotate_status, tagger_name, annotate_time, uid)
def update_policy(conn, info):
    sql = ''' UPDATE policy
              SET annotate_status = ? ,
                  tagger_name = ? ,
                  annotate_time = ?
              WHERE uid = ?'''
    cur = conn.cursor()
    cur.execute(sql, info)
    conn.commit()
```

```python
# 插入annotated_sentence表记录
# info = (uid, sid, sentence, sentence_type)
def insert_annotated_sentence(conn, info):
    sql = ''' INSERT INTO annotated_sentence(uid,sid,sentence,sentence_type)
              VALUES(?,?,?,?) '''
    cur = conn.cursor()
    cur.execute(sql, info)
    conn.commit()
```

```python
# 插入entity表记录
# info = (sid, entity, entity_type)
def insert_entity(conn, info):
    sql = ''' INSERT INTO entity(sid,entity,entity_type)
              VALUES(?,?,?) '''
    cur = conn.cursor()
    cur.execute(sql, info)
    conn.commit()
```

```python
# 插入entry表记录
# info = (sid, eid, entry_type, var, relation, field, var_context, field_context)
def insert_entry(conn, info):
    sql = ''' INSERT INTO entry(sid, eid, entry_type, var, relation, field, var_context, field_context)
              VALUES(?,?,?,?,?,?,?,?) '''
    cur = conn.cursor()
    cur.execute(sql, info)
    conn.commit()
```

```python
# 插入entry_logic表记录
# info = (uid, logic,logic_name)
def insert_entry_logic(conn, info):
    sql = ''' INSERT INTO entry_logic(uid, logic, logic_name)
              VALUES(?,?,?) '''
    cur = conn.cursor()
    cur.execute(sql, info)
    conn.commit()
```

```python
# 读取数据表记录
def read_table(conn, table_name):
    cursor = conn.cursor()
    data = cursor.execute("""select * from {}""".format(table_name))
    for d in data:
        print(d)
```

## 3 读文件、插记录

```python
import re
import glob
import numpy as np
import pandas as pd
```

```python
def read_file(conn, file_path):
#     file_path = '/Volumes/TOURO Mobil/policy标注/政策文件annotated/第二周/柏柯羽-第二周/503-柏柯羽.xlsx'
    path_items = re.split('[/.-]', file_path)
    tagger_name = path_items[-2]
    uid = str(int(path_items[-3]))
    annotate_time = path_items[-4]

    file_content = pd.read_excel(file_path).fillna('')
    sentence_num = file_content.shape[0]

    # 更新policy表中的文件状态
    update_policy(conn, (1, tagger, uid, annotate_time))
    # 对于每个句子
    for i in range(sentence_num):
        sid = uid + '_' + str(int(file_content.loc[i, '句子id']))
        # 插入到annotated_sentence表
        sentence = file_content.loc[i, '句子']
        sentence_type = file_content.loc[i, '句子类型']
        insert_annotated_sentence(conn, (uid, sid, sentence, sentence_type))
        # 插入到entity表
        for j in range(1, 6):
            if file_content.loc[i, '实体'+str(j)]!='' and file_content.loc[i, '实体类别'+str(j)]!='':
                insert_entity(conn, (sid, file_content.loc[i, '实体'+str(j)], file_content.loc[i, '实体类别'+str(j)]))
        # 插入到entry表
        for j in range(1, 16):
            if file_content.loc[i, '准入条件id'+str(j)]!='':
                insert_entry(conn, (sid, file_content.loc[i, '准入条件id'+str(j)], file_content.loc[i, '准入条件类别'+str(j)],
                                   file_content.loc[i, '变量'+str(j)], file_content.loc[i, '关系'+str(j)], file_content.loc[i, '域'+str(j)],
                                   file_content.loc[i, '变量{}的原文对应'.format(str(j))], file_content.loc[i, '域{}的原文对应'.format(str(j))]))
        # 插入到entry_logic表
        if file_content.loc[i, '准入条件逻辑关系']!='':
            insert_entry_logic(conn, (uid, file_content.loc[i, '准入条件逻辑关系'], file_content.loc[i, '准入条件认定组别']))
```

```python
total_folder_name = '/Volumes/TOURO Mobil/policy标注/政策文件annotated'
week_name = '第二周'
annotated_folder_list = glob.glob('{}/{}/*-{}'.format(total_folder_name, week_name, week_name))
```

```python
conn = sqlite3.connect('policy.db')
for annotated_folder in annotated_folder_list:
    annotated_files = glob.glob('{}/*.xlsx'.format(annotated_folder))
    for file in annotated_files:
        if '~' in file:
            continue
        print(file)
        read_file(conn, file)
conn.close()
```
